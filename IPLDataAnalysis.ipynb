```python
QUESTION:4 Match won by minimum wickets?
```

```python
min_wickets = matches[matches['win_by_wickets'] > 0]['win_by_wickets'].min()

match_min_wickets = matches[matches['win_by_wickets'] == min_wickets]

match_min_wickets
```

```python
INSIGHT FOR QUESTION:4
  * This match was won with the least number of wickets remaining.
  * The chasing team almost lost the match.
  * Such matches are very exciting for viewers.   
```

```python
QUESTION:5 Matches where D/L method was and wasn't applied?
```

```python
dl_matches = matches['dl_applied'].value_counts()

dl_matches
```

```python
sns.countplot(x='dl_applied', data=matches)

plt.title("D/L Method Applied or Not")

plt.show()
```

```python
INSIGHT FOR QUESTION:5
   * The dataset shows how many matches used the Duckworth-Lewis method.
   * The D/L method is applied when rain interrupts the match
   * Most IPL matches are completed without using this method.
```

```python
QUESTION:6 Number of matches held in each city? 
```

```python
city_matches = matches['city'].value_counts()

city_matches
```

```python
city_matches.head(10).plot(kind='bar')

plt.title("Matches in Different Cities")

plt.show()
```

```python
INSIGHT FOR QUESTION:6
  * Some cities host more IPL matches than others.
  * Major cities like Mumbai, Kolkata and Bangalore host many matches.
  * This is mainly due to larger stadiums and strong fan support.
```

```python
QUESTION:7 Number of matches won by each team?
```

```python
team_wins = matches['winner'].value_counts()

team_wins
```

```python
team_wins.plot(kind='bar')

plt.title("Matches Won by Teams")

plt.show()
```

```python
INSIGHT FOR QUESTION:7
  * The results show the total number of wins by each team.
  * Teams like Mumbai Indians and Chennai Super Kings have strong records.
  * Consistent performance helps teams win more matches.
```

```python
QUESTION:8 Number of matches held every season?
```

```python
season_matches = matches['season'].value_counts().sort_index()

season_matches
```

```python
sns.countplot(x='season', data=matches)

plt.xticks(rotation=90)

plt.show()
```

```python
INSIGHT FOR QUESTION:8
   * The number of matches varies across different seasons.
   * Some seasons have more matches due to more teams participating.
   * IPL has expanded over time, increasing the total matches.
```

```python
QUESTION:9 Top 10 players with most Man of the Match awards?
```

```python
QUESTION:10 Does winning the toss mean winning the match?
```

```python
toss_match_win = matches[matches['toss_winner'] == matches['winner']]

percentage = (len(toss_match_win) / len(matches)) * 100

percentage
```

```python
INSIGHT FOR QUESTION:10
   * The analysis shows the percentage of matches where toss winner also won the match.
   * Winning the toss gives a slight advantage.
   * However, match performance is more important than the toss.
```

```python
QUESTION:11 Decision taken after winning the toss?
```

```python
toss_decision = matches['toss_decision'].value_counts()

toss_decision
```

```python
INSIGHT FOR QUESTION:11
   * Most teams prefer to field first after winning the toss.
   * Chasing a target is often easier in T20 cricket.
   * Pitch conditions also influence this decision.
```
